```python
# Handeling by replacing those values using machine learning as imputing another value can bias the model at the end.

# Filter the original dataset to remove rows where total_stay is 0
filtered_df = hotels_df.copy()

# Processing
# Columns to one-hot encode
columns_to_encode = ['hotel', 'country', 'market_segment', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type']

def encoding(data, column):
    ''' this functions takes dataframe and column name/list to encode it(one-hot)'''

    data = pd.concat([data, pd.get_dummies(data[column], dtype=int, prefix=column)], axis=1)
    data = data.drop([column], axis=1)

    return data


for col in columns_to_encode:
    filtered_df = encoding(filtered_df, col)

# Display the encoded DataFrame
filtered_df.head()
```

```python
# Filtering the original dataset to remove rows where total_stay is 0
filtered_df_for_train = filtered_df[filtered_df['total_stays'] > 0]

X_weeknight_train = filtered_df_for_train.drop(['stays_in_weekend_nights', 'stays_in_week_nights', 'total_stays', 'arrival_date'], axis=1)[:100000]
y_weeknight_train = filtered_df_for_train['stays_in_week_nights'][:100000]
X_weeknight_test = filtered_df_for_train.drop(['stays_in_weekend_nights', 'stays_in_week_nights', 'total_stays', 'arrival_date'], axis=1)[100000: ]
y_weeknight_test = filtered_df_for_train['stays_in_week_nights'][100000: ]


X_weekend_train = filtered_df_for_train.drop(['stays_in_weekend_nights', 'stays_in_week_nights', 'total_stays', 'arrival_date'], axis=1)[:100000]
y_weekend_train = filtered_df_for_train['stays_in_weekend_nights'][:100000]
X_weekend_test = filtered_df_for_train.drop(['stays_in_weekend_nights', 'stays_in_week_nights', 'total_stays', 'arrival_date'], axis=1)[100000: ]
y_weekend_test = filtered_df_for_train['stays_in_weekend_nights'][100000: ]


# Defining the architecture of the neural network
model_weeknight = Sequential([
    Dense(64, activation='relu', input_shape=(X_weeknight_train.shape[1],)),
    Dropout(0.2),  # Add dropout for regularization
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1)  # Output layer with single neuron for regression
])

# Compiling the model
model_weeknight.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the model
history = model_weeknight.fit(X_weeknight_train, y_weeknight_train,
                    validation_data=(X_weeknight_test, y_weeknight_test),
                    epochs=10, batch_size=32, verbose=1)

# Evaluate the model
weeknight_loss = model_weeknight.evaluate(X_weeknight_test, y_weeknight_test, verbose=0)
print("Weeknight Model Loss:", weeknight_loss)


# Repeat the same process for the weekend model
model_weekend = Sequential([
    Dense(64, activation='relu', input_shape=(X_weekend_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1)
])

model_weekend.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

history = model_weekend.fit(X_weekend_train, y_weekend_train,
                    validation_data=(X_weekend_test, y_weekend_test),
                    epochs=10, batch_size=32, verbose=1)

weekend_loss = model_weekend.evaluate(X_weekend_test, y_weekend_test, verbose=0)
print("Weekend Model Loss:", weekend_loss)




# Impute values for missing rows in hotels_df_with_zeros using the trained models
hotels_df_with_zeros = filtered_df[filtered_df['total_stays'] < 1]

missing_data = hotels_df_with_zeros.drop(['stays_in_weekend_nights', 'stays_in_week_nights', 'total_stays', 'arrival_date'], axis=1)

# Predict using the trained neural network models
imputed_weeknight = model_weeknight.predict(missing_data)
imputed_weekend = model_weekend.predict(missing_data)

# Round up the predicted values to the nearest integer
imputed_weeknight_rounded = np.ceil(imputed_weeknight)
imputed_weekend_rounded = np.ceil(imputed_weekend)

# Replace the missing values in hotels_df_with_zeros with the rounded imputed values
hotels_df_with_zeros['stays_in_weekend_nights'] = imputed_weekend_rounded
hotels_df_with_zeros['stays_in_week_nights'] = imputed_weeknight_rounded

# Update the original dataframe (hotels_df) with the rounded imputed values
hotels_df.loc[hotels_df_with_zeros.index, 'stays_in_weekend_nights'] = imputed_weekend_rounded
hotels_df.loc[hotels_df_with_zeros.index, 'stays_in_week_nights'] = imputed_weeknight_rounded

# Adding stays_in_weekday_nights as column.
hotels_df['total_stays'] = hotels_df['stays_in_week_nights'] + hotels_df['stays_in_weekend_nights']
```

## 2. Hypothesis Testing

I will use a chi-square test to analyze the relationship between the two categorical variables: deposit_type and is_canceled.

Here's a step-by-step guide to perform the hypothesis test:

**Formulate Hypotheses:**
*  **Null Hypothesis (H0)**: There is no significant association between the type of deposit made and the likelihood of booking cancellation.
*  **Alternative Hypothesis (H1)**: The type of deposit made is associated with the likelihood of booking cancellation.


**Select Significance Level (α):**
I choose significance level of α = 0.05(which is very commonly used), or a 5% risk of concluding that an association exists when there is no actual association.

**Interpret Results:**
If the p-value is less than the chosen significance level α, we reject the null hypothesis and conclude that there is a significant association between deposit type and booking cancellation.

```python
# Writing Hypotheses:
H0 = "Null Hypothesis (H0): There is no significant association between the type of deposit made and the likelihood of booking cancellation."
H1 = "Alternative Hypothesis (H1): The type of deposit made is associated with the likelihood of booking cancellation."

# Defining the counts for each deposit type
no_deposit_count = df['deposit_type'].value_counts()[0]
non_refund_count = df['deposit_type'].value_counts()[1]
refundable_count = df['deposit_type'].value_counts()[2]


# Counting the number of canceled bookings for each deposit type
no_deposit_cancelled = df[(df['deposit_type'] == 'No Deposit') & (df['is_canceled'] == 1)].shape[0]
non_refund_cancelled = df[(df['deposit_type'] == 'Non Refund') & (df['is_canceled'] == 1)].shape[0]
refundable_cancelled = df[(df['deposit_type'] == 'Refundable') & (df['is_canceled'] == 1)].shape[0]

# Printing the counts
print("Number of bookings with No Deposit:", no_deposit_count)
print("Number of bookings with Non Refund:", non_refund_count)
print("Number of bookings with Refundable:", refundable_count)
print("Number of canceled bookings for No Deposit:", no_deposit_cancelled)
print("Number of canceled bookings for Non Refund:", non_refund_cancelled)
print("Number of canceled bookings for Refundable:", refundable_cancelled)


# Creating a contingency table using the provided counts
contingency_table = np.array([[no_deposit_count, non_refund_count, refundable_count],
                               [no_deposit_cancelled, non_refund_cancelled, refundable_cancelled]])

# Performing chi-square test for independence
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# I have used significance level to be 0.05 or 5%
alpha = 0.05

# Printing results
print("Chi-Square Statistic:", chi2)
print("P-value:", round(p_value, 6))
print("Degrees of Freedom:", dof)
print("Expected Frequencies:", expected)

# Interpreting results
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between deposit type and booking cancellation.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between deposit type and booking cancellation.")
```

**Conclusion:**
1. We Reject the null hypothesis, There is a significant association between **deposit type** and **booking cancellation**.

```python
def vif_calculator(df):
    '''Calculates VIF'''
    # Filter out non-numeric columns
    numeric_df = df.select_dtypes(include=['number'])

    # Calculate VIF for numeric columns
    vif_1 = pd.DataFrame()
    vif_1["variables"] = numeric_df.columns
    vif = [variance_inflation_factor(numeric_df.values, i) for i in range(numeric_df.shape[1])]
    vif_1["VIF"] = vif
    vif_1.sort_values(by='VIF', ascending=False, inplace=True)

    return vif_1
```

### Data Processor Class

```python
class DataPreprocessor:
    def __init__(self):
        self.encoder = None
        self.scaler = None

    def label_encode(self, df):
        # Label encoding for specific columns with dictionaries (if needed)
        label_encoding_hotel = {'Resort Hotel': 0, 'City Hotel': 1}
        df['hotel'] = df['hotel'].map(label_encoding_hotel)

        label_encoding_reserved = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H': 7, 'I': 8, 'K': 9, 'L': 10, 'P': 11}
        df['reserved_room_type'] = df['reserved_room_type'].map(label_encoding_reserved)

        label_encoding_assigned = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H': 7, 'I': 8, 'K': 9, 'L': 10, 'P': 11}
        df['assigned_room_type'] = df['assigned_room_type'].map(label_encoding_assigned)

        # One-hot encode the specified columns
        columns_one_hot = ['arrival_date_month', 'arrival_date_week_number', 'arrival_date_of_month', 'country',
                           'market_segment', 'deposit_type', 'customer_type']

        encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')  # Consider handle_unknown for unseen categories
        encoder.fit(df[columns_one_hot])
        encoded_df = pd.DataFrame(encoder.transform(df[columns_one_hot]),
                                  columns=encoder.get_feature_names_out(columns_one_hot))
        df = pd.concat([df.drop(columns_one_hot, axis=1), encoded_df], axis=1)

        self.encoder = encoder  # Save the fitted encoder object
        return df

    def save_encoder(self, filename):
        # Save the fitted encoder if it exists
        if self.encoder is not None:
            joblib.dump(self.encoder, filename)
            print(f"Encoder saved to: {filename}")
        else:
            print("Encoder not fitted yet. Cannot save.")

    def normalize(self, df):
        numerical_columns_to_scale = ['lead_time', 'stays_in_weekend_nights', 'adults', 'children',
                                      'previous_cancellations', 'previous_bookings_not_canceled',
                                      'stays_in_week_nights']

        df_features = df[numerical_columns_to_scale]
        scaler = MinMaxScaler()
        scaler.fit(df_features)
        df_features_scaled = pd.DataFrame(scaler.transform(df_features), columns=df_features.columns)
        df = df_features_scaled.join(df.drop(numerical_columns_to_scale, axis=1))
        self.scaler = scaler  # Save the fitted scaler object
        return df

    def save_scaler(self, filename):
        # Save the fitted scaler if it exists
        if self.scaler is not None:
            joblib.dump(self.scaler, filename)
            print(f"Scaler saved to: {filename}")
        else:
            print("Scaler not fitted yet. Cannot save.")
```

```python
class ModelTrainer:
    def __init__(self, X, y, task):
        self.X = X
        self.y = y
        self.task = task
        self.model = None
        self.history = None
        self.X_train, self.X_test, self.y_train, self.y_test = None, None, None, None

    def split_data(self, test_size=0.2, random_state=None):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=test_size, random_state=random_state)

    def create_model(self):
        if self.task == 'classify':
            # Define neural network for classification
            self.model = Sequential()
            self.model.add(Dense(units=64, activation='relu', input_dim=self.X.shape[1]))
            self.model.add(BatchNormalization())
            self.model.add(Dropout(0.2))  # Add dropout for regularization
            self.model.add(Dense(units=32, activation='relu'))
            self.model.add(BatchNormalization())
            self.model.add(Dropout(0.2))  # Add dropout for regularization
            self.model.add(Dense(units=1, activation='sigmoid'))  # Output layer with sigmoid for binary classification

        elif self.task == 'regression':
            # Define neural network for regression
            self.model = Sequential()
            self.model.add(Dense(units=128, activation='relu', input_dim=self.X.shape[1]))
            self.model.add(BatchNormalization())
            self.model.add(Dropout(0.2))  # Add dropout for regularization
            self.model.add(Dense(units=64, activation='relu'))  # Adjust hidden layer size as needed
            self.model.add(BatchNormalization())
            self.model.add(Dropout(0.2))  # Add dropout for regularization
            self.model.add(Dense(units=32, activation='relu'))
            self.model.add(Dense(units=1, activation='linear'))  # Output layer with linear activation for regression

        else:
            raise ValueError("Invalid value for 'task'. Choose either 'classify' or 'regression'.")

    def train_model(self, epochs= 30, batch_size=32):
        # Compile the model before training
        if self.task == 'classify':
            self.model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
        elif self.task == 'regression':
            self.model.compile(loss='mse', optimizer='adam', metrics=['mse', 'mae'])
        else:
            raise ValueError("Invalid value for 'task'. Choose either 'classify' or 'regression'.")

        # Train the model
        self.history = self.model.fit(self.X_train, self.y_train, validation_data=(self.X_test, self.y_test), epochs=epochs, batch_size=batch_size)

        return self.model

    def print_report(self):
        if self.task == 'classify':
            train_pred = self.model.predict(self.X_train)
            train_pred = (train_pred > 0.5).astype(int)
            print("Train Classification Report:")
            print(classification_report(self.y_train, train_pred))

            test_pred = self.model.predict(self.X_test)
            test_pred = (test_pred > 0.5).astype(int)
            print("\nTest Classification Report:")
            print(classification_report(self.y_test, test_pred))

        elif self.task == 'regression':
            train_pred = self.model.predict(self.X_train)
            print("Train Mean Squared Error:", mean_squared_error(self.y_train, train_pred))
            print("Train R^2 Score:", r2_score(self.y_train, train_pred))

            test_pred = self.model.predict(self.X_test)
            print("\nTest Mean Squared Error:", mean_squared_error(self.y_test, test_pred))
            print("Test R^2 Score:", r2_score(self.y_test, test_pred))

        else:
            raise ValueError("Invalid value for 'task'. Choose either 'classify' or 'regression'.")

    def print_learning_curve(self):
        if self.task == 'classify' and hasattr(self, 'history'):
            loss = self.history.history['loss']
            val_loss = self.history.history['val_loss']
            accuracy = self.history.history['accuracy']
            val_accuracy = self.history.history['val_accuracy']
            epochs = range(1, len(loss) + 1)

            plt.figure(figsize=(12, 6))

            # Plotting training and validation loss
            plt.subplot(1, 2, 1)
            plt.plot(epochs, loss, 'bo', label='Training loss')
            plt.plot(epochs, val_loss, 'b', label='Validation loss')
            plt.title('Training and Validation Loss')
            plt.xlabel('Epochs')
            plt.ylabel('Loss')
            plt.legend()

            # Plotting training and validation accuracy
            plt.subplot(1, 2, 2)
            plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
            plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
            plt.title('Training and Validation Accuracy')
            plt.xlabel('Epochs')
            plt.ylabel('Accuracy')
            plt.legend()

            plt.tight_layout()
            plt.show()

        elif self.task == 'regression' and hasattr(self, 'history'):
            loss = self.history.history['loss']
            val_loss = self.history.history['val_loss']
            mse = self.history.history['mse']
            val_mse = self.history.history['val_mse']
            mae = self.history.history['mae']
            val_mae = self.history.history['val_mae']
            epochs = range(1, len(loss) + 1)

            plt.figure(figsize=(12, 6))

            # Plotting training and validation loss
            plt.subplot(1, 2, 1)
            plt.plot(epochs, loss, 'bo', label='Training loss')
            plt.plot(epochs, val_loss, 'b', label='Validation loss')
            plt.title('Training and Validation Loss')
            plt.xlabel('Epochs')
            plt.ylabel('Loss')
            plt.legend()

            # Plotting training and validation metrics (MSE and MAE)
            plt.subplot(1, 2, 2)
            plt.plot(epochs, mse, 'bo', label='Training MSE')
            plt.plot(epochs, val_mse, 'b', label='Validation MSE')
            plt.plot(epochs, mae, 'go', label='Training MAE')
            plt.plot(epochs, val_mae, 'g', label='Validation MAE')
            plt.title('Training and Validation Metrics (MSE and MAE)')
            plt.xlabel('Epochs')
            plt.ylabel('Metrics')
            plt.legend()

            plt.tight_layout()
            plt.show()
        else:
            print("No training history available. Please train the model first.")


    def print_roc_auc_curve(self):
        if self.task == 'classify':
            # Predict probabilities
            y_pred_proba = self.model.predict(self.X_test)

            # Calculate fpr, tpr, and thresholds
            fpr, tpr, thresholds = roc_curve(self.y_test, y_pred_proba)

            # Calculate ROC AUC score
            roc_auc = auc(fpr, tpr)

            # Plot ROC curve
            plt.figure(figsize=(6,5))
            plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
            plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
            plt.xlim([0.0, 1.0])
            plt.ylim([0.0, 1.05])
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title('Receiver Operating Characteristic (ROC) Curve')
            plt.legend(loc="lower right")
            plt.show()

        else:
            print("ROC AUC curve is only applicable for classification tasks.")

    def reporter(self, train_report, test_report):
        if self.model is None:
            raise ValueError("Model not trained. Train the model first.")

        if self.task == 'classify':

            predecting_column = self.y.name

            # Extract the name of the model class
            model_name = "Neural Network"

            train_preds = self.model.predict(self.X_train)
            test_preds = self.model.predict(self.X_test)

            threshold = 0.5  # Adjust the threshold as needed
            train_preds = (train_preds > threshold).astype(int)
            test_preds = (test_preds > threshold).astype(int)

            train_cm = confusion_matrix(self.y_train, train_preds)
            test_cm = confusion_matrix(self.y_test, test_preds)

            train_TN, train_FN, train_TP, train_FP = train_cm[0][0], train_cm[1][0], train_cm[1][1], train_cm[0][1]
            test_TN, test_FN, test_TP, test_FP = test_cm[0][0], test_cm[1][0], test_cm[1][1], test_cm[0][1]

            train_specificity = train_TN / (train_TN + train_FP)
            test_specificity = test_TN / (test_TN + test_FP)

            train_acc = accuracy_score(self.y_train, train_preds)
            test_acc = accuracy_score(self.y_test, test_preds)

            train_roc = roc_auc_score(self.y_train, train_preds)
            test_roc = roc_auc_score(self.y_test, test_preds)

            train_prec = precision_score(self.y_train, train_preds)
            test_prec = precision_score(self.y_test, test_preds)

            train_rec = recall_score(self.y_train, train_preds)
            test_rec = recall_score(self.y_test, test_preds)

            train_f1 = f1_score(self.y_train, train_preds)
            test_f1 = f1_score(self.y_test, test_preds)

            train_row = pd.DataFrame({'feature': predecting_column, 'Model': model_name, 'Accuracy': train_acc, 'Precision': train_prec,
                            'Recall': train_rec, 'Specificity': train_specificity, 'F1 Score': train_f1, 'ROC': train_roc}, index=[0])
            test_row = pd.DataFrame({'feature': predecting_column, 'Model': model_name, 'Accuracy': test_acc, 'Precision': test_prec,
                            'Recall': test_rec, 'Specificity': test_specificity, 'F1 Score': test_f1, 'ROC': test_roc}, index=[0])

            train_report = pd.concat([train_report, train_row], ignore_index=True)
            test_report = pd.concat([test_report, test_row], ignore_index=True)

            return train_report, test_report


        else:
            print('Only for classification ')


    def reporter_regression(self, train_report_regression, test_report_regression):
        if self.model is None:
            raise ValueError("Model not trained. Train the model first.")

        if self.task == 'regression':

            predicting_column = self.y.name

            # Extract the name of the model class
            model_name = "Neural Network" + self.task

            train_preds = self.model.predict(self.X_train)
            test_preds = self.model.predict(self.X_test)

            train_mse = mean_squared_error(self.y_train, train_preds)
            test_mse = mean_squared_error(self.y_test, test_preds)

            train_rmse = np.sqrt(train_mse)
            test_rmse = np.sqrt(test_mse)

            train_r2 = r2_score(self.y_train, train_preds)
            test_r2 = r2_score(self.y_test, test_preds)

            n_train = len(self.y_train)
            n_test = len(self.y_test)
            k = self.X_train.shape[1]  # Number of predictors
            adj_train_r2 = 1 - ((1 - train_r2) * (n_train - 1) / (n_train - k - 1))
            adj_test_r2 = 1 - ((1 - test_r2) * (n_test - 1) / (n_test - k - 1))

            train_row = pd.DataFrame({'feature': predicting_column, 'Model': model_name, 'MSE': train_mse, 'RMSE': train_rmse,
                                      'R2 Score': train_r2, 'Adjusted R2': adj_train_r2}, index=[0])
            test_row = pd.DataFrame({'feature': predicting_column, 'Model': model_name, 'MSE': test_mse, 'RMSE': test_rmse,
                                    'R2 Score': test_r2, 'Adjusted R2': adj_test_r2}, index=[0])

            train_report_regression = pd.concat([train_report_regression, train_row], ignore_index=True)
            test_report_regression = pd.concat([test_report_regression, test_row], ignore_index=True)

            return train_report_regression, test_report_regression

        else:
            print('Only valid for regression')



    def save_model(self, model_name_given):
        # Get the name of the model class
        model_name = model_name_given

        # Generate the filename
        filename = f"Models\{model_name}.pkl"

        # Save the model to a pickle file
        with open(filename, 'wb') as file:
            pickle.dump(self.model, file)

        print(f"Model saved as {filename}")
```

```python
class random_forest_regressor:
    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.model = None
        self.history = None  # Corrected typo
        self.X_train, self.X_test, self.y_train, self.y_test = None, None, None, None

    def split_data(self, test_size=0.2, random_state=None):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=test_size, random_state=random_state)

    def train_model(self, random_state=42):
        # Initialize and train the Random Forest regressor
        self.model = RandomForestRegressor(random_state=random_state)
        self.history = self.model.fit(self.X_train, self.y_train)  # Store training history

        return self.model  # Return training history

    def evaluate_model(self):
        # Predictions on the testing set
        y_pred = self.model.predict(self.X_test)

        # Calculate Mean Squared Error and Mean Absolute Error
        mse = mean_squared_error(self.y_test, y_pred)
        mae = mean_absolute_error(self.y_test, y_pred)

        return mse, mae

    def plot_feature_importance(self, top_n=5):
        if self.model is not None:
            importance = self.model.feature_importances_
            indices = np.argsort(importance)[::-1]

            # Plot the feature importance
            plt.figure(figsize=(8, 6))
            plt.bar(range(top_n), importance[indices][:top_n], color='skyblue')
            plt.xticks(range(top_n), indices[:top_n], rotation=45, ha='right')
            plt.title('Top {} Feature Importance'.format(top_n))
            plt.xlabel('Feature')
            plt.ylabel('Importance')
            plt.tight_layout()
            plt.show()
        else:
            print("Model not trained. Train the model first.")


    def save_model(self, model_name_given):
        if self.model is not None:
            # Get the name of the model class
            model_name = model_name_given

            # Generate the filename
            filename = f"Models\{model_name}.pkl"

            # Save the model to a pickle file
            with open(filename, 'wb') as file:
                pickle.dump(self.model, file)

            print(f"Model saved as {filename}")
        else:
            print("Model not trained. Train the model first.")

    def reporter_regression(self, train_report_regression, test_report_regression):
        if self.model is None:
            raise ValueError("Model not trained. Train the model first.")

        predicting_column = self.y.name

        # Extract the name of the model class
        model_name = type(self.model).__name__

        train_preds = self.model.predict(self.X_train)
        test_preds = self.model.predict(self.X_test)

        train_mse = mean_squared_error(self.y_train, train_preds)
        test_mse = mean_squared_error(self.y_test, test_preds)

        train_rmse = np.sqrt(train_mse)
        test_rmse = np.sqrt(test_mse)

        train_r2 = r2_score(self.y_train, train_preds)
        test_r2 = r2_score(self.y_test, test_preds)

        n_train = len(self.y_train)
        n_test = len(self.y_test)
        k = self.X_train.shape[1]  # Number of predictors
        adj_train_r2 = 1 - ((1 - train_r2) * (n_train - 1) / (n_train - k - 1))
        adj_test_r2 = 1 - ((1 - test_r2) * (n_test - 1) / (n_test - k - 1))

        train_row = pd.DataFrame({'feature': predicting_column, 'Model': model_name, 'MSE': train_mse, 'RMSE': train_rmse,
                                  'R2 Score': train_r2, 'Adjusted R2': adj_train_r2}, index=[0])
        test_row = pd.DataFrame({'feature': predicting_column, 'Model': model_name, 'MSE': test_mse, 'RMSE': test_rmse,
                                'R2 Score': test_r2, 'Adjusted R2': adj_test_r2}, index=[0])

        train_report_regression = pd.concat([train_report_regression, train_row], ignore_index=True)
        test_report_regression = pd.concat([test_report_regression, test_row], ignore_index=True)

        return train_report_regression, test_report_regression
```

```python
class XGBModelAnalysis:
    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.x_train = None
        self.x_test = None
        self.y_train = None
        self.y_test = None
        self.model = None


    def split_data(self, test_size=0.2, random_state=None):
        self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(self.X, self.y,
                                                                                test_size=test_size,
                                                                                random_state=random_state,
                                                                                stratify=self.y)
    def create_model(self, **kwargs):
        self.model = xgb.XGBClassifier(objective='multi:softmax', **kwargs)

    def train_model(self):
        self.model.fit(self.x_train, self.y_train)

        return self.model

    def evaluate_model(self):
        # Evaluate on test data
        test_preds = self.model.predict(self.x_test)
        test_accuracy = accuracy_score(self.y_test, test_preds)
        test_cm = confusion_matrix(self.y_test, test_preds)
        test_cls_report = classification_report(self.y_test, test_preds)

        # Evaluate on train data
        train_preds = self.model.predict(self.x_train)
        train_accuracy = accuracy_score(self.y_train, train_preds)
        train_cm = confusion_matrix(self.y_train, train_preds)
        train_cls_report = classification_report(self.y_train, train_preds)

        print("-------------------- Evaluation Report for Test Data --------------------")
        print("Accuracy:", test_accuracy)
        print("Confusion Matrix:\n", test_cm)
        print("Classification Report:\n", test_cls_report)

        print("-------------------- Evaluation Report for Train Data --------------------")
        print("Accuracy:", train_accuracy)
        print("Confusion Matrix:\n", train_cm)
        print("Classification Report:\n", train_cls_report)

    def get_best_params(self):
        if self.best_params:
            return self.best_params
        else:
            raise ValueError("Best params are not available. Run grid_search method first.")

    def plot_feature_importance(self, top_n=5):
        if self.model_name in ["DecisionTreeClassifier", "RandomForestClassifier", "GradientBoostingClassifier"]:
            importance = self.model.feature_importances_
            feature_names = self.X.columns.tolist()
            indices = np.argsort(importance)[::-1]

            plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
            plt.bar(range(top_n), importance[indices][:top_n], color='skyblue')
            plt.xticks(range(top_n), [feature_names[i] for i in indices][:top_n], rotation=45, ha='right')
            plt.title('Top {} Feature Importance'.format(top_n))
            plt.xlabel('Feature')
            plt.ylabel('Importance')
            plt.tight_layout()
            plt.show()
        else:
            print("Feature importance is not available for", self.model_name)


    def reporter(self, train_report, test_report):
        if self.model is None:
            raise ValueError("Model not trained. Train the model first.")

        predicting_column = self.y_train.name
        model_name = type(self.model).__name__

        train_preds = self.model.predict(self.x_train)
        test_preds = self.model.predict(self.x_test)

        print("Shape of train_preds:", train_preds.shape)
        print("Shape of test_preds:", test_preds.shape)

        train_cm = confusion_matrix(self.y_train, train_preds)
        test_cm = confusion_matrix(self.y_test, test_preds)

        num_classes = 12  # Number of classes

        print("Number of classes:", num_classes)
        print("Shape of train_cm:", train_cm.shape)
        print("Shape of test_cm:", test_cm.shape)

        # Check if the confusion matrix shape matches the number of classes
        if train_cm.shape[0] != num_classes or test_cm.shape[0] != num_classes:
            raise ValueError("Number of classes in confusion matrix does not match the expected number of classes.")

        train_specificity = []
        test_specificity = []
        for i in range(num_classes):
            train_TN = np.sum(np.delete(train_cm[i], i))
            train_FP = np.sum(np.delete(train_cm[:, i], i))
            train_specificity.append(train_TN / (train_TN + train_FP))

            test_TN = np.sum(np.delete(test_cm[i], i))
            test_FP = np.sum(np.delete(test_cm[:, i], i))
            test_specificity.append(test_TN / (test_TN + test_FP))

        train_acc = accuracy_score(self.y_train, train_preds)
        test_acc = accuracy_score(self.y_test, test_preds)

        train_prec = precision_score(self.y_train, train_preds, average='weighted')
        test_prec = precision_score(self.y_test, test_preds, average='weighted')

        train_rec = recall_score(self.y_train, train_preds, average='weighted')
        test_rec = recall_score(self.y_test, test_preds, average='weighted')

        train_f1 = f1_score(self.y_train, train_preds, average='weighted')
        test_f1 = f1_score(self.y_test, test_preds, average='weighted')

        train_roc = roc_auc_score(label_binarize(self.y_train, classes=np.unique(self.y_train)),
                                  self.model.predict_proba(self.x_train),
                                  average='macro')
        test_roc = roc_auc_score(label_binarize(self.y_test, classes=np.unique(self.y_test)),
                                self.model.predict_proba(self.x_test),
                                average='macro')

        train_row = pd.DataFrame({'feature': predicting_column, 'Model': model_name, 'Accuracy': train_acc,
                                  'Precision': train_prec, 'Recall': train_rec, 'Specificity': "Nan",
                                  'F1 Score': train_f1, 'ROC': train_roc}, index=[0])

        test_row = pd.DataFrame({'feature': predicting_column, 'Model': model_name, 'Accuracy': test_acc,
                                'Precision': test_prec, 'Recall': test_rec, 'Specificity': "Nan",
                                'F1 Score': test_f1, 'ROC': test_roc}, index=[0])

        train_report = pd.concat([train_report, train_row], ignore_index=True)
        test_report = pd.concat([test_report, test_row], ignore_index=True)

        return train_report, test_report


    def save_model(self, model_name_given):
        # Get the name of the model class
        model_name = model_name_given

        # Generate the filename
        filename = f"Models\{model_name}.pkl"

        # Save the model to a pickle file
        with open(filename, 'wb') as file:
            pickle.dump(self.model, file)

        print(f"Model saved as {filename}")

```

### Clustering Classes

```python

class AutoClusterAnalysis:
    def __init__(self, X):
        self.X = X
        self.kmeans_model = None
        self.hierarchical_model = None
        self.dbscan_model = None
        self.kmeans_clusters = None
        self.hierarchical_clusters = None
        self.dbscan_clusters = None
        self.best_kmeans_n_clusters = None
        self.best_hierarchical_n_clusters = None
        self.best_dbscan_eps = None

    def find_best_kmeans_clusters(self, n_cluster_range):
        silhouette_scores = []
        for n_clusters in n_cluster_range:
            kmeans = KMeans(n_clusters=n_clusters)
            cluster_labels = kmeans.fit_predict(self.X)
            silhouette_avg = silhouette_score(self.X, cluster_labels)
            silhouette_scores.append(silhouette_avg)

        self.best_kmeans_n_clusters = n_cluster_range[np.argmax(silhouette_scores)]
        return self.best_kmeans_n_clusters

    def find_best_hierarchical_clusters(self, n_cluster_range):
        silhouette_scores = []
        for n_clusters in n_cluster_range:
            hierarchical = AgglomerativeClustering(n_clusters=n_clusters)
            cluster_labels = hierarchical.fit_predict(self.X)
            silhouette_avg = silhouette_score(self.X, cluster_labels)
            silhouette_scores.append(silhouette_avg)

        self.best_hierarchical_n_clusters = n_cluster_range[np.argmax(silhouette_scores)]
        return self.best_hierarchical_n_clusters

    def find_best_dbscan_eps(self, eps_range):
        silhouette_scores = []
        for eps in eps_range:
            dbscan = DBSCAN(eps=eps)
            cluster_labels = dbscan.fit_predict(self.X)
            silhouette_avg = silhouette_score(self.X, cluster_labels)
            silhouette_scores.append(silhouette_avg)

        self.best_dbscan_eps = eps_range[np.argmax(silhouette_scores)]
        return self.best_dbscan_eps

    def fit_kmeans(self, n_clusters):
        self.kmeans_model = KMeans(n_clusters=n_clusters)
        self.kmeans_clusters = self.kmeans_model.fit_predict(self.X)
        return self.kmeans_clusters

    def fit_hierarchical(self, n_clusters):
        self.hierarchical_model = AgglomerativeClustering(n_clusters=n_clusters)
        self.hierarchical_clusters = self.hierarchical_model.fit_predict(self.X)
        return self.hierarchical_clusters

    def fit_dbscan(self, eps):
        self.dbscan_model = DBSCAN(eps=eps)
        self.dbscan_clusters = self.dbscan_model.fit_predict(self.X)
        return self.dbscan_clusters
```

```python
# Creating a copy for rfm alanysis
rfm_df = hotels_df.copy()

# Adding new column called total_previous_bookings
rfm_df['total_previous_bookings'] = rfm_df['previous_bookings_not_canceled'] + rfm_df['previous_cancellations']


# Adding new column called days_since_last_booking
# Converting 'arrival_date' column to datetime format
rfm_df['arrival_date'] = pd.to_datetime(rfm_df['arrival_date'])

# Finding the last date available in the dataset because i will consider the last date of booking from that day
last_date = rfm_df['arrival_date'].max()

# Calculating the number of days since the last booking for each row
rfm_df['days_since_last_booking'] = (last_date - rfm_df['arrival_date']).dt.days


# Converint reserved_room_type to number
label_encoding_reserved = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H': 7, 'I': 8, 'K': 9, 'L': 10, 'P': 11}
rfm_df['reserved_room_type'] = rfm_df['reserved_room_type'].map(label_encoding_reserved)



# Only taking required columns
rfm_df = rfm_df[['days_since_last_booking' , 'previous_cancellations', 'total_previous_bookings', 'reserved_room_type']]

# Checking all columns
rfm_df.head()
```

### K-means Clustring

#### 1. Finding Optimum Numbers of Clusters using Elbow Method

```python
# Convert rfm_pca to a NumPy array if it's a DataFrame
rfm_pca = rfm_pca.values if isinstance(rfm_pca, pd.DataFrame) else rfm_pca

# 1. Finding the best optimal cluster number using the elbow method
list_1 = []

for k in range(1, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(rfm_pca)
    list_1.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 20), list_1, "-o")
plt.grid(True)
plt.xlabel("Number of Clusters (k)", fontsize=12)
plt.ylabel("Sum of Square Distances", fontsize=12)
plt.title("Elbow Method For Optimal k", fontsize=14)
plt.xticks(range(1, 20))
plt.show()
```

### Comprehensive Cluster Evaluation and Ranking

#### Studying characteristics of Each cluster

```python
# Function to calculate mean of values in a column grouped by cluster number
def calculate_cluster_means(dataframe, column):
    # Group by cluster number and calculate the mean for the specified column
    cluster_means = dataframe.groupby('cluster_number')[column].mean()
    # Sort the mean values in descending order
    cluster_means_sorted = cluster_means.sort_values(ascending=False)
    print(column)
    print(cluster_means_sorted)
    return cluster_means_sorted
```

```python
# For studying only
rfm_df_study = rfm_df.copy()
rfm_df_study['cluster_number'] = kmeans.labels_
```

```python
rfm_df_study.head()
```

```python
# Initializing the dictionary to store cluster rankings
cluster_rankings = {
    "days_since_last_booking": [],
    "previous_cancellations": [],
    "total_previous_bookings": [],
    "reserved_room_type": []
}

# Iterating through each column in the DataFrame
for column in rfm_df.columns:
    # Calculating mean values for each cluster in the current column
    cluster_means = calculate_cluster_means(rfm_df_study, column)

    # Sorting clusters based on mean values in descending order
    sorted_clusters = sorted(cluster_means.items(), key=lambda x: x[1], reverse=True)

    # Reversing the order of ranks for specific columns
    if column in ["days_since_last_booking", "previous_cancellations"]:
        sorted_clusters = sorted(sorted_clusters, key=lambda x: x[1])

    # Extractiing cluster numbers and store them in the dictionary
    cluster_numbers = [cluster for cluster, _ in sorted_clusters]
    cluster_rankings[column] = cluster_numbers

# Printing the cluster rankings dictionary
print("Cluster Rankings:")
for column, rankings in cluster_rankings.items():
    print(f"For {column} ranks from best cluster to worst: {rankings}")
```
